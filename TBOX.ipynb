```python
from rdflib import Graph, Namespace, RDF, RDFS

g = Graph()
pub = Namespace("http://www.example.org/publication#")
g.bind("pub", pub)

classes = ["Abstract","Author","City","Event","Conference","Workshop","Editor",
           "ConferenceChair","JournalEditor","Reviewer","HIndex","Journal",
           "Volume","Proceedings","Paper","Keyword","Year"]
for c in classes:
    g.add((pub[c], RDF.type, RDFS.Class))

g.add((pub.Conference,      RDFS.subClassOf, pub.Event))
g.add((pub.Workshop,        RDFS.subClassOf, pub.Event))
g.add((pub.Editor,          RDFS.subClassOf, pub.Author))
g.add((pub.ConferenceChair, RDFS.subClassOf, pub.Editor))
g.add((pub.JournalEditor,   RDFS.subClassOf, pub.Editor))
g.add((pub.Reviewer,        RDFS.subClassOf, pub.Author))

def prop(uri, dom, ran):
    g.add((uri, RDF.type, RDF.Property))
    g.add((uri, RDFS.domain, dom))
    g.add((uri, RDFS.range,  ran))

prop(pub.assignedBy,       pub.Reviewer,      pub.Editor)
prop(pub.belongsTo,        pub.Volume,        pub.Journal)
prop(pub.cites,            pub.Paper,         pub.Paper)
prop(pub.contains,         pub.Paper,         pub.Abstract)
prop(pub.edits,            pub.JournalEditor, pub.Journal)
prop(pub.hasIndex,         pub.Author,        pub.HIndex)
prop(pub.isCorrespondingAuthor, pub.Author,   pub.Paper)
prop(pub.isAbout,          pub.Paper,         pub.Keyword)
prop(pub.isPartOf,         pub.Proceedings,   pub.Event)
prop(pub.ofVenueIn,        pub.Proceedings,   pub.City)
prop(pub.presentedIn,      pub.Paper,         pub.Proceedings)
prop(pub.presentedInYear,  pub.Proceedings,   pub.Year)
prop(pub.publishedIn,      pub.Paper,         pub.Volume)
prop(pub.publishedInYear,  pub.Volume,        pub.Year)
prop(pub.reviews,          pub.Reviewer,      pub.Paper)
prop(pub.runs,             pub.ConferenceChair, pub.Event)
prop(pub.writes,           pub.Author,        pub.Paper)

g.serialize("publication_TBOX.ttl", format="turtle")
print("T-Box written to publication_TBOX.ttl")
```
